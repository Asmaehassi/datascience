```python
!pip install ucimlrepo
import matplotlib.pyplot as plt
import seaborn as sns

# Re-fetch dataset to ensure X and y are defined for plotting
from ucimlrepo import fetch_ucirepo
#  TÃ©lÃ©chargement de la base de donnÃ©es "Adult" depuis UCI
adult = fetch_ucirepo(id=2)

# Extraction des donnÃ©es
# X contient les variables explicatives (Ã¢ge, Ã©ducation, profession, etc.)
# y contient la variable cible (le revenu : >50K ou <=50K)
X = adult.data.features
y = adult.data.targets

# On combine X et y dans un seul DataFrame pour faciliter la visualisation
df = X.copy()
df['income'] = y


# Visualisation 1 : distribution de la variable 'income'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='income', palette='viridis')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

```python
#  Visualisation 2 : distribution de l'Ã¢ge selon le niveau de revenu
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='income', kde=True, palette='magma')
plt.title('Distribution of Age by Income')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Distribution of 'hours-per-week' by 'income'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='hours-per-week', hue='income', kde=True, palette='cividis')
plt.title('Distribution of Hours-per-Week by Income')
plt.xlabel('Hours-per-Week')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Distribution of 'education' by 'income'
plt.figure(figsize=(12, 7))
sns.countplot(data=df, y='education', hue='income', palette='plasma', order=df['education'].value_counts().index)
plt.title('Distribution of Education Level by Income')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.tight_layout()
plt.show()
```
